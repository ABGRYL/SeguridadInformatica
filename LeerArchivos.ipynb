```python
Los modos comunes son:

"r": Lectura (por defecto).
"w": Escritura (sobrescribe el archivo si ya existe).
"a": Añadir (agrega contenido al final del archivo si ya existe).
"b": Modo binario (se utiliza junto con otros modos, como "rb" o "wb").
```

```python
import PyPDF2
with open('proyecto.pdf', 'rb') as pdf_file:

    pdf_reader = PyPDF2.PdfReader(pdf_file)

    # Accede al número de páginas en el archivo PDF
    num_pages = len(pdf_reader.pages)
    print(f'Número de páginas en el PDF: {num_pages}')
    for page_number in range(2):  
        page = pdf_reader.pages[page_number]
        page_text = page.extract_text()
        print(f'Contenido de la página {page_number + 1}:\n{page_text}\n')
```

```python
pdf_file.close()
```

```python
import docx

doc = docx.Document('proyecto.docx')

# Lee el contenido del documento
document_text = []

for paragraph in doc.paragraphs:
    document_text.append(paragraph.text)
full_text = '\n'.join(document_text)

print(full_text)
```
